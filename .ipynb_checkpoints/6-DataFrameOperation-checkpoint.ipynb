### Working with missing data

```python
weather_na_df = pd.read_excel('6-weatherna.xlsx')
```

```python
weather_na_df.isna()
```

```python
weather_na_df.describe() # boş datalar olsa da hesaplamaya devam eder.
```

```python
weather_na_df.dropna() # rowun herhangi bir columnsunda null değer varsa o rowu siler.
```

```python
weather_na_df.dropna(axis=1) #null değer olan columnsu siler ama tercih edilen bi yöntem değildir genellikle bütün veri silinir.
```

```python
weather_na_df.drop('Paris', axis = 1) # genellikle en çok null değer olan columnsu silmek için o columns belirtlierek yapılır.
```

```python
weather_na_df.fillna(20) #bütün boşlukları verilen değerle doldurur
```

```python
weather_na_df
```

```python
weather_na_df.fillna(weather_na_df.mean()) # her kolonu kendi ortalaması ile doldurur.
```

### Multi Index

```python
first_index = ["Got","Got","Got","Hotd","Hotd","Hotd"]
```

```python
inner_index = ["Dany","Jon","Arya","Daemon","Rhenrya","Viserys"]
```

```python
zipped_index= list(zip(first_index,inner_index))
```

```python
zipped_index
```

```python
zipped_index = pd.MultiIndex.from_tuples(zipped_index)
```

```python
zipped_index
```

```python
sample_index = np.ones((6,2))
```

```python
last_frame = pd.DataFrame(sample_index , index=zipped_index, columns=["Age","Salary"])
```

```python
last_frame
```

```python
last_frame.loc["Got"].loc["Dany"] #multydata olduğunda indexlere ulaşmak için loc fonksiyonu bu şekilde kullanılır.
```

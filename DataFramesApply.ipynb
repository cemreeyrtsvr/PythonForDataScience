```python
import pandas as pd
```

```python
df= pd.read_csv('8-apply_function_data.csv')
```

```python
df
```

```python
def maaş(salary):
    if salary<50000:
        return "Low"
    elif 50000<= salary <80000:
        return "Medium"
    else:
        return "High"
```

```python
df["Maaş Deneyimi"]=df["Salary"].apply(maaş) #fonksiyonları frame yansıtmayı sağlar
```
